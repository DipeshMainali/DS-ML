### Build a K-NN model to predict survival


### 1. Download the data and build a pandas DataFrame

```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_data = pd.read_csv("data/train.csv")
```

```python
x_df = train_data.drop(columns=['Survived'])
y_df = train_data[['Survived']]
```

### 2. Split the data into testing and training and testing sets

```python
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df)
```

### 3. Change the data types of columns, if necessary

```python
for column in train_data.columns:
    if train_data[column].dtype == type(object):
        le = sklearn.preprocessing.LabelEncoder()
        train_data[column] = le.fit_transform(train_data[column])

train_data.fillna(0, inplace=True)
train_data
```

### 4. Build a KNN models with different numbers of neighbors.  What is the best accuracy you get (on testing set) ?

```python
x_df = train_data.drop(columns=['Survived'])
y_df = train_data['Survived']
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, random_state=555)
for i in range(1, 20):
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(x_train, y_train)
    print('Accuracy at {}: {:.2f}'.format(i, clf.score(x_test, y_test)))

print('Best Accuracy (accounting performance) is at {}: {:.2f}'.format(5, 0.70))
```

### 5. What accuracy would you get if you accidentally tested your accuracy on the TRAINING set ?

```python
new_clf = KNeighborsClassifier(n_neighbors=5)
new_clf.fit(x_train, y_train)
print('Accuracy: {:.2f}'.format(new_clf.score(x_train, y_train)))
```

### 6. Define euclidean distance.

Euclidean distance is the length of the line segment between 2 points in a euclidean space (multi-dimension space)

Mathematically,

The ***Euclidean distance*** between two points ***\((x_1, x_2, ..., x_n)\)*** and ***\((y_1, y_2, ..., y_n)\)*** in an n-dimensional space is calculated as:

$$
d = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$
