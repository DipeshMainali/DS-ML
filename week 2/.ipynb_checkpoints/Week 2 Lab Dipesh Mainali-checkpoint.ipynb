### 1. Create four DataFrames from the four .csv files

```python
df_Q1 = pd.read_csv("data/Divvy_Trips_2018_Q1.csv")
df_Q2 = pd.read_csv("data/Divvy_Trips_2018_Q2.csv")
df_Q3 = pd.read_csv("data/Divvy_Trips_2018_Q3.csv")
df_Q4 = pd.read_csv("data/Divvy_Trips_2018_Q4.csv")
df_Q3
```

### 2. Create a single dataframe by combining the four previously created

### 11. Create a function that calculates age.  Use the difference between the year of the activity (can use datetime library for this) and the birthyear column.

hint: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html

```python
# def cal_age(travel_year, dob):
    # return travel_year - dob

def cal_age(data):
    return data['travel_year'] - data['birthyear']
```

### 12. Use the %time magic to rerun your calculation for #11.  How long it takes to run it?

```python
%%time
df['travel_year'] = df['start_time'].dt.year
applied_value = df.apply(cal_age, axis=1)
# df.apply(lambda row: cal_age(df['travel_year'], df['birthyear']), axis=1)
```
