### 1. Retrieve the Kaggle Chicago Bike Dataset we used in Week 2's lab
https://www.kaggle.com/michaelshoemaker/divvy-bike-chicago-2018

```python
import pandas as pd
df_Q1 = pd.read_csv("data/Divvy_Trips_2018_Q1.csv")
df_Q2 = pd.read_csv("data/Divvy_Trips_2018_Q2.csv")
df_Q3 = pd.read_csv("data/Divvy_Trips_2018_Q3.csv")
df_Q4 = pd.read_csv("data/Divvy_Trips_2018_Q4.csv")
```

### 2. Combine the data into one dataframe if you haven't already

```python
df = pd.concat([df_Q1, df_Q2, df_Q3, df_Q4])
df
```

### 3. Calculate the mean trip duration by gender.  You may first have to convert the tripduration column.

hint: https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html  
errors='coerce'

```python
df['tripduration'] = pd.to_numeric(df['tripduration'].str.replace(',', ''), errors='coerce')
```

```python
df.groupby('gender').agg({'tripduration': 'mean'})
```
